```python
import re
import pandas as pd
```

```python
Mydf = pd.DataFrame(columns=["grant_id","patent_title","kind","number_of_claims","inventors","citations_applicant_count","citations_examiner_count","claims_text","abstract"])
```

```python
fData = open('Group126.txt','r')
rData=fData.read()
outData=open('Finaloutput1.json','a+')
contents=rData.split('<?xml version="1.0" encoding="UTF-8"?>')
```

```python
outGrantid=""
outPatenttitle=""
outPrint=""
outClaimcount=""
countApplicant=0
countExaminer=0
stringClaimtext=""
stringAbstract=""
jsonContent=""

for lines in contents:
    
    grantId = re.compile(r'file="(\w+\d+)-(.*?).XML"')
    matchGrantid=grantId.search(lines)
    if matchGrantid != None:
        outGrantid=matchGrantid.group(1)
    
    
        patentTitle = re.compile(r'<invention-title id(.*?)>(.*?)<\/(.*?)>')
        matchPatenttitle=patentTitle.search(lines)
        if matchPatenttitle:
            outPatenttitle=matchPatenttitle.group(2)
    
    
        appReference=re.compile(r'<application-reference appl-type="(.*?)">')
        matchAppreference=appReference.search(lines)
        if matchAppreference:
            capture=matchAppreference.group(1)   
            if(capture=="utility"):
                kindPattern=re.compile(r'<kind>(.*?)<\/\w*>')
                findKind=kindPattern.findall(lines)
                kindMatch=findKind[0]
                if(kindMatch=="B1"):
                    outPrint="Utility Patent Grant (no pre-grant publication) issued on or after January 2, 2001."
                else:
                    outPrint="Utility Patent Grant (with pre-grant publication) issued on or after January 2, 2001."
            elif(capture=="plant"):
                kindPattern=re.compile(r'<kind>(.*?)<\/\w*>')
                findKind=kindPattern.findall(lines)
                kindMatch=findKind[0]
                if(kindMatch=="P2"):
                    outPrint="Plant Patent Grant (no pre-grant publication) issued on or after January 2, 2001."
                else:
                    outPrint="Plant Patent Grant (with pre-grant publication) issued on or after January 2, 2001."
            elif(capture=="design"):
                outPrint="Design Patent"
            else:
                outPrint="Reissue Patent"
    
    
    
        claimCount = re.compile(r'<number-of-claims>(.*?)<\/number-of-claims>')
        matchClaimcount=claimCount.search(lines)
        if matchClaimcount:
            outClaimcount = matchClaimcount.group(1)
    
    
    
        inventorsPattern = re.compile(r'<inventors>(.*?)</inventors>',re.DOTALL)
        matchInventors = inventorsPattern.findall(lines)
        for each in matchInventors:
            name1 = re.compile(r'(?:<last-name>)(.*?)(?:</last-name>)',re.DOTALL)
            lastName = name1.findall(each)
            name2 = re.compile(r'(?:<first-name>)(.*?)(?:</first-name>)',re.DOTALL)
            firstName = name2.findall(each)
            i = 0
            inventorName = []
            for each in firstName:  
                fullName=""
                fullName= firstName[i] + " " + lastName[i] 
                inventorName.append(fullName)
                i+=1    
            inventorName=('[%s]' % ', '.join(map(str, inventorName)))
        
        applicantCitedpattern= re.compile(r'(?:<category>)(cited by applicant)(?:<\/\w+>)')
        matchApplicantcited = applicantCitedpattern.findall(lines)
        countApplicant=len(matchApplicantcited)
    
    
    
        examinerCitedpattern = re.compile(r'<category>(cited by examiner)<\/\w*>')
        matchExaminercited = examinerCitedpattern.findall(lines)
        countExaminer=len(matchExaminercited)
    
    
    
        claimTextpattern = re.compile(r'<claims id="claims">(.*?)</claims>',re.DOTALL)
        matchClaimtext = claimTextpattern.findall(lines)
        for each in matchClaimtext:
            stringClaimtext = (' ,').join(matchClaimtext)
            stringClaimtext = re.sub(r'<(.*?)>','',stringClaimtext)
            stringClaimtext = re.sub('\s+',' ', stringClaimtext)
            stringClaimtext = ("["+stringClaimtext+"]")
    
        abstractPattern = re.compile(r'<abstract id="abstract">(.*?)</abstract>', re.DOTALL)
        matchAbstract = abstractPattern.findall(lines)
        if(len(matchAbstract)!=0):
            stringAbstract = (', ').join(matchAbstract)
            stringAbstract = re.sub(r'<(.*?)>','',stringAbstract)
            stringAbstract = re.sub('\s+', ' ', stringAbstract)
        else:
            stringAbstract="NA"
    
    
        Mydf = Mydf.append({"grant_id":outGrantid,"patent_title":outPatenttitle,"kind":outPrint,"number_of_claims":outClaimcount,"inventors":inventorName,"citations_applicant_count":countApplicant,"citations_examiner_count":countExaminer,"claims_text":stringClaimtext,"abstract":stringAbstract},ignore_index=True)
        j=0
        
        jsonContent=('{"'+outGrantid+'":{"patent_title":"'+outPatenttitle+'","kind":"'+outPrint+'","number_of_claims":'+str(outClaimcount)+',"inventors":"'+inventorName+'","citations_applicant_count":'+str(countApplicant)+',"citations_examiner_count":'+str(countExaminer)+',"claims_text":"'+stringClaimtext+'","abstract":"'+stringAbstract+'"}')
        outData.write(jsonContent)
        
    else:
        outData.write("")
        
```

```python
Mydf.to_csv('Finaloutput.csv', index = False, encoding='utf-8')
```

```python
outData.close()
fData.close()
```
